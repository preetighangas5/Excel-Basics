**Question 1**: Explaint the difference between Relative, Absolute and Mixed Cell Referencing ?
**Answer:** ->* Relative referencing* changes automatically when formulas are copied to another cell,maintaining the relative position between the cells.For example : If cell C1 contains=A1+B1 and you copu it down to C2, the formula automatically adjusts to =A2+B2.
*An absolute reference* fixed a specific cell in place so it never changes, regardless of where the formula is copied.Use dollar($)sign before the row and column to get it fixed.Example: If cell C1 contains=$A$1*B1 and you copy it down to C2, the first part remains $A$1 while the B1 changes to B2.
*Mixed cell referencing* fixes only one part of the cell address either the row and column while leaving the other relative.For example : Locked Row = A$1
          Locked Column= $A1

**Question 2:** You have the following dataset
               Vehicle        Price
              
               Car            500000
               Bicycle         10500
               Auto rikshaw    25000
  Write a formula to calculate the total sales of Car and Bicycle only.
 ** Answer ** : **bold text**=SUMIF(A2:A4, "Car", B2:B4) + SUMIF(A2:A4, "bicycle",B2:B4)

 Question 3 : Using the data below, write a formula to calculate the average sales of items priced above 100 but less than 300

      ITEM    PRICE    SALES
       A       90       1000
       B       150      1200
       C       250      1500
      
Answer: =AVERAGEIFS(C2:C4,B2:B4,">100",B2:B4,"<300")
